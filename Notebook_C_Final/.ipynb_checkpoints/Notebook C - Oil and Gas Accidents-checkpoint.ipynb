# Notebook C: Oil and Gas Pipeline Accidents in Kern County

This notebook plots hazardous liquid and natural gas accidents in oil and gas pipelines in Kern County since 2010. I use national data from the US Department of Transportation: Pipeline and Hazardous Materials Safety Administration (PHMSA) and clip that data to Kern County using subsetted CalEnviroScreen data. This data illustrates further how oil and gas wells are important polluting variables but that the associated infrastructure, such as pipelines, pose their own hazards. Reading briefly into PHMSA, it seems that this data is likely incomplete given chronic underfunding at the agency.   

```python
import pandas as pd
import geopandas as gpd
import folium
import contextily as cx
import matplotlib.pyplot as plt
```

Here, I read the hazardous liquid accident data, display the columns and subset them to only include spillage data, company name, year, and geographic coordinates.

```python
haz = pd.read_csv('accident_hazardous_liquid_jan2010_present.csv', low_memory=False)
```

```python
pd.set_option('display.max_columns', None)
haz.head()
```

```python
hazsub = haz[['LOCATION_LATITUDE','LOCATION_LONGITUDE', 'UNINTENTIONAL_RELEASE_BBLS','COMMODITY_RELEASED_TYPE','IYEAR','NAME']]
```

```python
hazsub.head()
```

```python
hazsub = gpd.GeoDataFrame(
    hazsub, geometry = gpd.points_from_xy(hazsub.LOCATION_LONGITUDE,hazsub.LOCATION_LATITUDE),crs = "EPSG:4326")
```

```python
hazsub.plot()
```

I read CalEnviroScreen data which I then subset to only include Kern County, and then further subset to only include geometry data, tract data, and pollution burden score. I project the coordinates to match those of the hazardous liquid data. 

```python
gdf = gpd.read_file('CES4 Final Shapefile.shp',
                   dtype= {
        'Tract':str,
    }, crs = "EPSG:4326")
```

```python
gdf_kern = gdf.loc[gdf['County'] == 'Kern'] #subsets data to just include Kern County
gdf_kern.sample()
```

```python
gdf_kern = gdf_kern.to_crs("EPSG:4326")
```

```python
#subsetting data further by column 
#'Tract','TotPop19', 'PesticideP','Tox_Rel_P', 'DrinkWat', 
                #'DrinkWatP', 'Lead_P', 'CleanupP', 'GWThreatP', 
                #    'HazWasteP', 'ImpWatBod', 'ImpWatBodP', 'SolWasteP', 
                 #     'PolBurdSc', 'PolBurdP', 'Asthma', 
                  #   'AsthmaP', 'LowBirWP','CardiovasP',
       #  'Poverty','Hispanic', 'White', 'AfricanAm',
        #             'NativeAm', 'OtherMult',
gdf_ksub = gdf_kern[['PolBurdSc', 'Tract', 'Shape_Leng', 'Shape_Area', 'geometry']]
gdf_ksub.sample(1)
```

I clip the hazardous waste data to the Kern County tract data and then convert years to string given past issues with the way the year variable showed up in tool tips. 

```python
haz_clip = gpd.clip(hazsub, gdf_ksub)
```

```python
haz_clip.dtypes
```

```python
haz_clip["IYEAR"]= haz_clip["IYEAR"].apply(str)
```

```python
haz_clip.dtypes
```

```python
haz_clip.shape
```

I read the natural gas accident from the same data source and over the same time period. I then follow the same process as above by subsetting, converting to a geodataframe, and clipping to the Kern County boundaries.   

```python
gas = pd.read_csv('incident_gas_distribution_jan2010_present.csv')
```

```python
pd.set_option('display.max_columns', None)
gas.head()
```

```python
gas = gas[['LOCATION_LATITUDE','LOCATION_LONGITUDE','COMMODITY_RELEASED_TYPE','UNINTENTIONAL_RELEASE','IYEAR','NAME']]
```

```python
gas.head()
```

```python
gas = gpd.GeoDataFrame(
    gas, geometry = gpd.points_from_xy(gas.LOCATION_LONGITUDE,gas.LOCATION_LATITUDE),crs = "EPSG:4326")
```

```python
gas_clip = gpd.clip(gas, gdf_ksub)
gas_clip.head(10)
```

```python
gas_clip["IYEAR"] = gas_clip["IYEAR"].apply(str)
```

Here, I map the pollution burden which provides context for the distribution of pollution more broadly in the county, over which I add the hazardous liquid data and natural gas accident data. I use tool tips so users can see which company owns the pipelines that have had accidents, in what year they occured, what amount of oil or gas was leaked, and the kind of commodity released. 

We can see that for the hazardous liquid data that the substance appears to be always crude oil which makes sense given the level of extraction in that area. Additionally, it is notable that there are multiples owners of pipeline infrastructure perhaps adding a layer of difficulty in regulating multiple companies. 

```python
Kern2 = folium.Map(location=[35.4,-118.8], 
               zoom_start = 8,
               tiles='CartoDB positron', 
               attribution='CartoDB')
  
style1 = {'fillColor': '#000000', 'color': '#000000', 'weight': 1.5}
style2 = {'fillColor': '#7491FF', 'color': '#7491FF',  "weight": 1.5}
style3 = {'fillColor': '#B01043', 'color': '#B01043',  "weight": 1.5}
style4 = {'fillColor': '#DE76F5', 'color': '#DE76F5',  "weight": 1.5}


folium.Choropleth(
    geo_data=gdf_ksub,
    name="Census Tracts",
    data=gdf_ksub,
    columns=["Tract", "PolBurdSc"],
    key_on="feature.properties.Tract", #this gets into geojson structure of data ie feature.properties.column name
    #for each tract map the pollution score
    fill_color="YlGn",
    fill_opacity=0.5,
    line_opacity=0.2,
    line_weight = 0.8,
    edgecolor = '#5A5A5A',
    legend_name="Pollution Burden Score",
).add_to(Kern2)

folium.GeoJson(
    data=gas_clip, 
    name="Natural Gas Accidents", 
    style_function=lambda x: {'color':'black','fillColor':'transparent','weight':0.5},
    tooltip=folium.features.GeoJsonTooltip(
                        fields=[
                                'NAME', 'IYEAR', 'COMMODITY_RELEASED_TYPE', 'UNINTENTIONAL_RELEASE'
                               ],
                        aliases=[
                                 "Name:", 'Year', 'Commodity Released', 'Volume of gas released in thousand standard cubic feet', 
                                ], 
                        localize=True,
                        sticky=False,
                        labels=True,
                        style="""
                            background-color: #F0EFEF;
                            border: 2px solid black;
                            border-radius: 3px;
                            box-shadow: 3px;
                        """,
                        max_width=800,),
                            highlight_function=lambda x: {'weight':3,'fillColor':'grey'},
).add_to(Kern2)


folium.GeoJson(
    data=haz_clip, 
    name="Hazardous Liquid Accidents", 
    style_function=lambda x: {'color':'black','fillColor':'transparent','weight':0.5},
    tooltip=folium.features.GeoJsonTooltip(
                        fields=[
                                'NAME', 'IYEAR', 'COMMODITY_RELEASED_TYPE', 'UNINTENTIONAL_RELEASE_BBLS'
                               ],
                        aliases=[
                                 "Name:", 'Year', 'Commodity Released', 'Release of oil (in barrels)',
                                ], 
                        localize=True,
                        sticky=False,
                        labels=True,
                        style="""
                            background-color: #F0EFEF;
                            border: 2px solid black;
                            border-radius: 3px;
                            box-shadow: 3px;
                        """,
                        max_width=800,),
                            highlight_function=lambda x: {'weight':3,'fillColor':'grey'},
 ).add_to(Kern2) 
folium.LayerControl().add_to(Kern2)

Kern2.save("final_accidents_.html")

Kern2
```

```python
gas_clip['color'] = ''
```

```python
haz_clip['color'] = ''
```

```python
gas_clip.loc[gas_clip['COMMODITY_RELEASED_TYPE'] == 'NATURAL GAS', 'color'] = 'blue'
```

```python
haz_clip.loc[haz_clip['COMMODITY_RELEASED_TYPE'] == 'CRUDE OIL', 'color'] = 'black'
```

```python
Kern5 = folium.Map(location=[35.4,-118.8], 
               zoom_start = 8,
               tiles='CartoDB positron', 
               attribution='CartoDB')


folium.Choropleth(
    geo_data=gdf_ksub,
    name="Census Tracts",
    data=gdf_ksub,
    columns=["Tract", "PolBurdSc"],
    key_on="feature.properties.Tract", #this gets into geojson structure of data ie feature.properties.column name
    #for each tract map the pollution score
    fill_color="YlGn",
    fill_opacity=0.5,
    line_opacity=0.2,
    line_weight = 0.8,
    edgecolor = '#5A5A5A',
    legend_name="Pollution Burden Score",
).add_to(Kern5)

for index, row in gas_clip.iterrows():
    name="Natural Gas Accidents", 
    tooltip_text = ' Company: '+ row.NAME +  '----'+ 'Commodity released: '  + row.COMMODITY_RELEASED_TYPE + '----'+ row.IYEAR + 'Year: ' + row.IYEAR
    folium.Marker(
        [row.LOCATION_LATITUDE,row.LOCATION_LONGITUDE], 
        popup=row.IYEAR, 
        tooltip=tooltip_text,
        legend=True,
        icon=folium.Icon(color=row.color)
    ).add_to(Kern5)

for index, row in haz_clip.iterrows():
    name="Crude Oil Accidents", 
    tooltip_text = ' Company: '+ row.NAME +  '----'+ 'Commodity released: '  + row.COMMODITY_RELEASED_TYPE + '----'+ 'Year: ' + row.IYEAR
    folium.Marker(
        [row.LOCATION_LATITUDE,row.LOCATION_LONGITUDE], 
        popup=row.IYEAR, 
        tooltip=tooltip_text,
        icon=folium.Icon(color=row.color)
    ).add_to(Kern5)
Kern5.save("final_accidents_marker.html")

# show the map
Kern5

```

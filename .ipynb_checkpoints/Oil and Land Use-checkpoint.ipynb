```python
oil = pd.read_csv('data/kern_county_oil.csv')
oil.head(10)
```

```python
oil.columns.to_list()
```

```python
oilsubset = oil[['WellStatus', 'OperatorNa', 'WellTypeLa','Latitude','Longitude','WellSymbol']]
oilsubset.head()
```

```python
oilsubset = oilsubset[oilsubset.WellStatus != 'Canceled']
```

```python
oilsubset = oilsubset[oilsubset.WellStatus != 'Unknown']
```

```python
oilsubset = oilsubset[oilsubset.WellStatus != 'Plugged']
```

```python
oilsubset = oilsubset[oilsubset.WellStatus != 'PluggedOnly']
```

```python
oilsubset = oilsubset.loc[oilsubset['WellTypeLa'] == 'Oil & Gas'] #subsets to just oil and gas data
```

```python
well_count = oilsubset.value_counts(['WellStatus']).reset_index(name="count")
well_count
```

```python
gdf = gpd.read_file('data/CES4 Final Shapefile.shp',
                   dtype= {
        'Tract':str,
    }, crs = "EPSG:3857")
```

```python
gdf_kern = gdf.loc[gdf['County'] == 'Kern'] #subsets data to just include Kern County
gdf_kern.sample()
```

```python
gdf_kern.dtypes
```

```python
gdf_kern['Tract'] = gdf_kern['Tract'].astype(np.str)
gdf_kern.dtypes
```

```python
gdf_kern.sample()
```

```python
gdf_kern['Tract'] = gdf_kern['Tract'].str.zfill(11)
gdf_kern.sample()
```

```python
gdf_kern.plot(figsize=(10,10))
```

```python
#subsetting data further by column 
#'Tract','TotPop19', 'PesticideP','Tox_Rel_P', 'DrinkWat', 
                #'DrinkWatP', 'Lead_P', 'CleanupP', 'GWThreatP', 
                #    'HazWasteP', 'ImpWatBod', 'ImpWatBodP', 'SolWasteP', 
                 #     'PolBurdSc', 'PolBurdP', 'Asthma', 
                  #   'AsthmaP', 'LowBirWP','CardiovasP',
       #  'Poverty','Hispanic', 'White', 'AfricanAm',
        #             'NativeAm', 'OtherMult',
gdf_ksub = gdf_kern[['TotPop19', 'GWThreatP', 'DrinkWatP', 'PolBurdSc', 'Tract', 'Shape_Leng', 'Shape_Area', 'geometry']]
gdf_ksub.sample(1)
```

```python
gdf_ksub.crs
```

```python
#reading file with crs 4326 didn't work so this step ensured it changed
gdf_ksub = gdf_ksub.to_crs("EPSG:3857") 
gdf_ksub.crs
```

```python
censuscrop = gpd.sjoin(crops, gdf_ksub, how="right") 
censuscrop.shape
```

```python
censuscrop.head()
```

```python
censuscrop.plot(figsize=(10,10))
```

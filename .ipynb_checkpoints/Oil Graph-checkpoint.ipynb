```python
import pandas as pd
import numpy as np
import seaborn
import folium
```

```python
import geopandas as gpd
```

```python
import contextily as cx
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
oil = pd.read_csv('data/kern_county_oil.csv')
oil.head(10)
```

```python
oil.columns.to_list()
```

```python
oilsubset = oil[['WellStatus', 'WellTypeLa','Latitude','Longitude']]
oilsubset.head()
```

```python
oilsubset = oilsubset[oilsubset.WellStatus != 'Canceled']
```

```python
oilsubset = oilsubset[oilsubset.WellStatus != 'Unknown']
```

```python
oilsubset = oilsubset[oilsubset.WellStatus != 'Plugged']
```

```python
oilsubset = oilsubset[oilsubset.WellStatus != 'PluggedOnly']
```

```python
oilsubset = oilsubset.loc[oilsubset['WellTypeLa'] == 'Oil & Gas'] #subsets to just oil and gas data
```

```python
oilsubset.sample(10)
oilsubset.shape
```

```python
gdfoil = gpd.GeoDataFrame(
    oilsubset, geometry = gpd.points_from_xy(oilsubset.Longitude,oilsubset.Latitude),crs = "EPSG:4326")
```

```python
gdfoil.plot()
```

```python
gdf = gpd.read_file('data/CES4 Final Shapefile.shp',
                   dtype= {
        'Tract':str,
    }, crs = "EPSG:4326")
```

```python
gdf_kern = gdf.loc[gdf['County'] == 'Kern'] #subsets data to just include Kern County
gdf_kern.sample()
```

```python
gdf_kern.dtypes
```

```python
gdf_kern['Tract'] = gdf_kern['Tract'].astype(np.str)
gdf_kern.dtypes
```

```python
gdf_kern.sample()
```

```python
gdf_kern['Tract'] = gdf_kern['Tract'].str.zfill(11)
gdf_kern.sample()
```

```python
gdf_kern.plot(figsize=(10,10))
```

```python
gdf_kern.columns.to_list()
```

```python
#subsetting data further by column 
#'Tract','TotPop19', 'PesticideP','Tox_Rel_P', 'DrinkWat', 
                #'DrinkWatP', 'Lead_P', 'CleanupP', 'GWThreatP', 
                #    'HazWasteP', 'ImpWatBod', 'ImpWatBodP', 'SolWasteP', 
                 #     'PolBurdSc', 'PolBurdP', 'Asthma', 
                  #   'AsthmaP', 'LowBirWP','CardiovasP',
       #  'Poverty','Hispanic', 'White', 'AfricanAm',
        #             'NativeAm', 'OtherMult',
gdf_ksub = gdf_kern[['County','Asthma','AsthmaP','Cardiovas','CardiovasP','LowBirtWt',
 'LowBirWP','Pesticide','PM2_5','TotPop19', 'ImpWatBod','ImpWatBodP', 'GWThreat','GWThreatP', 'DrinkWat', 'DrinkWatP', 'PolBurdSc', 'Tract', 'Shape_Leng', 'Shape_Area', 'geometry']]
gdf_ksub.sample(1)
```

```python
#reading file with crs 4326 didn't work so this step ensured it changed
gdf_ksub = gdf_ksub.to_crs("EPSG:4326") 
```

```python
# yk: let's make sure both datasets align
fig,ax = plt.subplots(figsize=(10,10))
gdf_ksub.plot(ax=ax,facecolor='#ccc',edgecolor='white',lw=0.2)
gdfoil.plot(ax=ax,color='red',markersize=20,marker='x',alpha=0.5)
```

```python
gdfoil.shape
```

```python
gdf_ksub.shape
```

```python
# yk: changed this so that we assign a census tract number to each oil well
oilcalmerge = gpd.sjoin(gdfoil, gdf_ksub,how="left")
oilcalmerge.shape
```

```python
# yk: do a value count of tracts to see how many per tract
oil_count = oilcalmerge.value_counts('Tract').reset_index(name='count')
oil_count.head(20)
```

```python
# yk: merge this back to the census tract geojson
oilcalmerge = gdf_ksub.merge(oil_count, how='left', on='Tract')
oilcalmerge.head(1)
```

```python
#this allows NA tracts with no oil wells to appear as 0's on the legend 
oilcalmerge["count"] = oilcalmerge["count"].fillna(0)
```

```python
oilcalmerge.tail(20)
```

```python
# yk: reproject to mercator
oilcalmerge_mercator=oilcalmerge.to_crs(3857)
```

```python
gdfoil_mercator=gdfoil.to_crs(3857)
```

```python
oilcalmerge_mercator.dtypes
```

```python
#converting count to integer from float so legend will plot better
oilcalmerge_mercator['count'] = oilcalmerge_mercator['count'].astype(int)
```

```python
oilcalmerge_mercator.head(4)
```

```python
# yk: map with two layers
fig,ax = plt.subplots(figsize=(12,4))

# background census tracts to see the outlines
#oilcalmerge_mercator.plot(ax=ax,alpha=0.6,edgecolor='#bbb',facecolor='black', legend=True) 
#above code was needed before converting NAs to zeros so 0 would be included on the legend 

# census tracts colored by count
oilcalmerge_mercator.plot(ax=ax,column='count',edgecolor='#bbb',  cmap='Reds', alpha=0.8,
                    legend=True)

ax.axis('off')
cx.add_basemap(ax=ax, source=cx.providers.CartoDB.Positron)
```

```python
#still some NAs...and ordering is not working
oilcalmerge_corr = oilcalmerge_corr.sort_values(by='count', ascending=False)
oilcalmerge.head()
```

```python

a=oilcalmerge_corr['count']
e=oilcalmerge_corr['PolBurdSc']
c=np.array(oilcalmerge_corr['count'])
d=np.array(oilcalmerge_corr['PolBurdSc'])
m, b = np.polyfit(c, d, 1)
```

```python
#use population density - possibly a relationship between pop density and oil wells 
plt.plot(a,e,'o')
plt.plot(c, m*c + b)
plt.title(label="Oil and Pollution Burden",
          fontsize=20,
          color="black",)
plt.xlabel('Oil & Gas Wells per Census Tract', fontsize=12)
plt.ylabel('Pollution Burden Score', fontsize=12)
plt.tight_layout()
plt.subplots_adjust(top=0.85)
```
